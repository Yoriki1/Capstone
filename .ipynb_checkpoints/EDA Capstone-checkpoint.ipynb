```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dji = pd.read_csv('dji_clean')
```

```python
dji.info()
```

```python
dji['date'] = pd.to_datetime(dji['date'])
```

```python
dji = dji.set_index('date')
```

```python
dji.head()
```

```python
dji_new = pd.DataFrame(dji, columns = ['stock', 'close', 'open', 'high', 'low', 'volume', 'percent_change_next_weeks_price'])
```

```python
AA = dji_new.query('stock == "AA"')
AXP = dji_new.query('stock == "AXP"')
BA = dji_new.query('stock == "BA"')
BAC = dji_new.query('stock == "BAC"')
CAT = dji_new.query('stock == "CAT"')
CSCO = dji_new.query('stock == "CSCO"')
CVX = dji_new.query('stock == "CVX"')
DD = dji_new.query('stock == "DD"')
DIS = dji_new.query('stock == "DIS"')
GE = dji_new.query('stock == "GE"')
HD = dji_new.query('stock == "HD"')
HPQ = dji_new.query('stock == "HPQ"')
IBM = dji_new.query('stock == "IBM"')
INTC = dji_new.query('stock == "INTC"')
JNJ = dji_new.query('stock == "JNJ"')
JPM = dji_new.query('stock == "JPM"')
KO = dji_new.query('stock == "KO"')
KRFT = dji_new.query('stock == "KRFT"')
MCD = dji_new.query('stock == "MCD"')
MMM = dji_new.query('stock == "MMM"')
MRK = dji_new.query('stock == "MRK"')
MSFT = dji_new.query('stock == "MSFT"')
PFE = dji_new.query('stock == "PFE"')
PG = dji_new.query('stock == "PG"')
T = dji_new.query('stock == "T"')
TRV = dji_new.query('stock == "TRV"')
UTX = dji_new.query('stock == "UTX"')
VZ = dji_new.query('stock == "VZ"')
WMT = dji_new.query('stock == "WMT"')
XOM = dji_new.query('stock == "XOM"')
```

```python
dji_30 = [AA, AXP, BA, BAC, CAT, CSCO, CVX, DD, DIS, GE ,HD, HPQ, IBM, INTC, JNJ, JPM, KO, KRFT, MCD, MMM, MRK, MSFT, PFE, PG, T, TRV, UTX, VZ, WMT, XOM]
dji_30_tick = ['AA', 'AXP', 'BA', 'BAC', 'CAT', 'CSCO', 'CVX', 'DD', 'DIS', 'GE' ,'HD', 'HPQ', 'IBM', 'INTC', 'JNJ', 'JPM', 'KO', 'KRFT', 'MCD', 'MMM', 'MRK', 'MSFT', 'PFE', 'PG', 'T', 'TRV', 'UTX', 'VZ', 'WMT', 'XOM']
```

```python
plt.figure(figsize=(20, 20))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(dji_30, 1):
    plt.subplot(10, 3, i)
    company['close'].plot()
    plt.ylabel('close')
    plt.xlabel(None)
    plt.title(f"Closing Price of {dji_30_tick[i - 1]}")
    
    
plt.tight_layout()
```

```python
dji.hist('close')
plt.xlabel('Closing')
plt.ylabel('Count')
plt.title(f"Closing Price Distribution")
```

```python
dji_new.sort_values('close')
```

```python
plt.figure(figsize=(20, 20))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(dji_30, 1):
    plt.subplot(10, 3, i)
    company['volume'].plot()
    plt.ylabel('volume')
    plt.xlabel(None)
    plt.title(f"Volume of {dji_30_tick[i - 1]}")
    
    
plt.tight_layout()
```
